####  1.0 Access and analyse Sea Surface Tempearture 

#### 1.1 Load data from the National Data Buoy Center (NBDC)

#### 1.1.1 Import working python packages

```python
# Notebook Setup
import xarray as xr
## !pip install netcdf4
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = xr.open_dataset('https://dods.ndbc.noaa.gov/thredds/dodsC/data/stdmet/acyn4/acyn4.ncml')
data1 = data.sel(time=slice('2020-06-01','2020-06-08'))
data2 = data.sel(time=slice('2020-06-01','2020-06-30'))
data3 = data.sel(time=slice('2019-06-30','2020-06-30'))
fldata = xr.open_dataset('https://dods.ndbc.noaa.gov/thredds/dodsC/data/stdmet/kywf1/kywf1.ncml')
```

#### 1.2 Visualizing SST 

#### 1.2.2 One Week

```python
data1.sea_surface_temperature.plot(color ='r')
plt.title('Sea surface temperature over the first week of June 2020 off the coast of Atlantic City, NJ');
plt.xlabel('Time');
plt.ylabel('Sea surface temperature (C)');
```

#### 1.2.4 Comparing 2 Locations
